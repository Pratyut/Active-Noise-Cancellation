# AutoCorrelation of Arbitrary Signal Shapes

#### Importing Modules

```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
delta = 50  # Set step size here
```

```python
def Autointeracplot(x1,x2,x3,x4,x5,x6,x7,x8):
    
  
    y1 = np.linspace(x1, x2, delta)
    y2 = np.linspace(x2, x3, delta)
    y3 = np.linspace(x3, x4, delta)
    y4 = np.linspace(x4, x5, delta)
    y5 = np.linspace(x5, x6, delta)
    y6 = np.linspace(x6, x7, delta)
    y7 = np.linspace(x7, x8, delta)
    y  = np.concatenate((y1,y2,y3,y4,y5,y6,y7))
    
       
    plt.figure(1, figsize=(12, 3), dpi=80, facecolor='w', edgecolor='k')
    plt.plot(y,linewidth=1,c='b')
    plt.title("Signal Synthesised")
    plt.axvline(0,linewidth=3,c='r')
    plt.ylim([-2, 6])
    plt.grid(1)
    
    z = np.correlate(y,y,"full")
    plt.figure(2, figsize=(12, 3), dpi=80, facecolor='w', edgecolor='k')
    plt.title("AutoCorrelation of Synthesised Signal")
    plt.plot(z)
        
```

```python
from IPython.html.widgets import interact
```

```python
interact(Autointeracplot, x1 =2, x2 =2, x3 = 2, x4 = 2, x5 = 2, x6 = 2, x7 = 2, x8 = 2)
```
