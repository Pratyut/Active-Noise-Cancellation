####Lets construct a one dimensional signal of sinusoidal shape. This signal changes its value spatialy as well as with respect to time. So the signal y is a function of space and time.

\begin{align}
y(x,t) & = f(x-ct)\\
\end{align}

####Import Modules

```python
import math
import sympy
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

####Parameter Specifications

```python
delta = 1000        # Set step size here
A     = 1;          # Set wave amplitude here   
f     = 1000;       # Ser frequency of basic wave here
c     = 343.0;      # Speed of sound in air
t     = 0.0;        # Select the time instant
lamda = c/f;        # Find the wavelength
```

####Signal Generation

```python
x = np.arange(0, lamda, (1.0/delta) )
y = A*np.sin((2*np.pi/lamda)*x-(c*t));
```

####Pad it with zeros for better visual appeal

```python
L     = len(y)
#pad  = np.zeros((L,))
#y    = np.concatenate((pad,y,pad), axis=1)
```

####Plot the resulting signal

```python
plt.plot(y)
```

\begin{align}
y(x,t) & = f(x-ct)\\
\end{align}

\begin{align}
y(x,t) & = f(x-f \lambda t)\\
\end{align}

\begin{align}
y(x,t) & = Acos(\frac{2\pi}{\lambda}(x-\frac{\lambda t}{T})) \\
\end{align}

\begin{align}
y(x,t) & = Acos(\frac{2\pi x}{\lambda}-\frac{2\pi t}{T})) \\
\end{align}

####Now its a sinusoid of Amplitude A, Frequency f, Velocity c...etc. Its a function of space and time.

```python
def OneDplot(t):
    
    delta = 100      # Set step size here
    A     = 1;        # Set wave amplitude here   
    f     = 1000;     # Ser frequency of basic wave here
    c     = 343.0;    # Speed of sound in air
    
    lamda = c/f;      # Find the wavelength 
    x = np.arange(0, lamda, (1.0/delta) )
    y = A*np.sin((2*np.pi/lamda)*x-(c*t));
    
    
    L    = len(y)
    #pad = np.zeros((L,))
    #y = np.concatenate((pad,y,pad), axis=1)
    plt.plot(y)
    plt.title('1D Wave')
    plt.xlabel('Points in Space')
    plt.ylabel('Amplitue')
    plt.grid(1)
    
```

```python
from IPython.html.widgets import interact
```

```python
interact(OneDplot, t= (0.0,0.01,(1.0/delta)))
```

#### Clearly its a right going wave. As time progress the wave propagates to the positive x or to right side 

#### Similarly we can have a left going wave. As time progress the wave propagates to the negative x or to the left side 

```python
def OneDplotR(t):
    
    delta = 100      # Set step size here
    A     = 1;        # Set wave amplitude here   
    f     = 1000;     # Ser frequency of basic wave here
    c     = 343.0;    # Speed of sound in air
    
    lamda = c/f;      # Find the wavelength 
    x = np.arange(0, lamda, (1.0/delta) )
    y = A*np.sin((2*np.pi/lamda)*x+(c*t));
    
    
    L    = len(y)
    #pad = np.zeros((L,))
    #y = np.concatenate((pad,y,pad), axis=1)
    plt.plot(y)
    plt.title('1D Wave')
    plt.xlabel('Points in Space')
    plt.ylabel('Amplitue')
    plt.grid(1)    
```

```python
from IPython.html.widgets import interact
```

```python
interact(OneDplotR, t= (0.0,0.01,(1.0/delta)))
```

### Lets try standing waves

```python
def standingwave(t):
    
    delta = 10000      # Set step size here
    A     = 1;        # Set wave amplitude here   
    f     = 1000;     # Ser frequency of basic wave here
    c     = 343.0;    # Speed of sound in air
    
    lamda = c/f;      # Find the wavelength 
    x = np.arange(0, lamda, (1.0/delta) )
    y1 = A*np.sin((2*np.pi/lamda)*x-(c*t));
    y2 = A*np.sin((2*np.pi/lamda)*x+(c*t));
    y=y1+y2
    
    L    = len(y)
    #pad = np.zeros((L,))
    #y = np.concatenate((pad,y,pad), axis=1)
    plt.plot(y)
    plt.ylim([-2.0,2.0])
    plt.title('1D Wave')
    plt.xlabel('Points in Space')
    plt.ylabel('Amplitue')
    plt.grid(1)    
```

```python
interact(standingwave, t= (0.0,0.01,(1.0/delta)))
```
