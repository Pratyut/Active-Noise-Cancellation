##Plotting an arbitrary sin wave:

Parameter Specifications:

```python
Fs  = 8000  # Set Sampling Rate here
f   = 1000  # Set frequency here
T   = 1     # Set duration of wave to be generated here
```

Generate Time Vector:

```python
t= np.arange(0, T, (1.0/Fs) )
```

Generate Output Vector

```python
x = np.sin(2 * np.pi * f * t)
```

And plot!

```python
rangeToPlot = arange(0, Fs/100)

plot(t[rangeToPlot], x[rangeToPlot])
title('Generated Wave')
xlabel('Time(in sec)')
ylabel('Voltage(V)')

grid(1)
```

Write to wav and play:

```python
from scikits.audiolab import wavwrite

#Normalize + Export:
xNorm = x/amax(abs(x))
wavwrite(xNorm, 'sinWave' + '.wav', Fs)
```

And play in notebook:

```python
from IPython.display import Audio
Audio('sinWave' + '.wav')
```

##Controlling frequency with a slider:

```python
def generateSineWave(frequency):
    Fs  = 8000  # Set Sampling Rate here
    T   = 1     # Set duration of wave to be generated here
    t= np.arange(0, T, (1.0/Fs) )
    
    x = np.sin(2 * np.pi * frequency * t)
    
    rangeToPlot = arange(0, Fs/100)

    plot(t[rangeToPlot], x[rangeToPlot])
    title('Generated Wave')
    xlabel('Time(in sec)')
    ylabel('Voltage(V)')

    grid(1)
    
```

```python
generateSineWave(1000)
```

```python
from IPython.html.widgets import interact
```

```python
interact(generateSineWave, frequency = (100, 4000))
```
