#Spot the Pi


Import libaries

```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Create and plot the basic plot

```python
delta = 60  # Set step size here
```

```python
x= np.linspace(-1, 1, delta)
padder= np.linspace(0,0,delta)
```

```python
y=np.sqrt(1-x**2)
Ly=len(y)
```

```python
plt.plot(x, y, linewidth=2)
plt.grid(1)
```

Pad its edges with zeros so that we can shift

```python
ynew=np.lib.pad(y, (delta,delta), 'constant')
```

```python
xnew=np.arange(0, 3*Ly,1)-(Ly*1.5)
```

```python
plt.plot(xnew,ynew,linewidth=2,c='g')
plt.grid(1)
plt.ylim([0, 4])
```

Choose a costant x value and observe it y intercept as we shift

```python
def shiftplot(shift):
    yshifted=np.roll(ynew,shift)
    
    plt.figure(1, figsize=(16, 3), dpi=80, facecolor='w', edgecolor='k')
    
    
    plt.plot(xnew,ynew,linewidth=1,c='b')

    plt.plot(xnew,yshifted,linewidth=3,c='g')
    plt.axvline(0,linewidth=3,c='r')
    plt.scatter(0, yshifted[3*Ly/2], c='b',linewidth=3)
    plt.ylim([0, 1.1])
    plt.xlim([-(1.5*delta), (1.5*delta)])
    plt.grid(1)    
```

```python
from IPython.html.widgets import interact
```

```python
interact(shiftplot, shift= (-60, 60,1))
```

The vertical line is sweeping through the wave. So when there are many secondary waves its essentially summing up all the points, very similar to integration. Its like summing all the y values of different secondary waves for any given fixed x value.

So wave front amplitude at any point  in x axis will be sum of all elements of fundamental wave

We intuitively know area of half circle is pi/2 or we can calculate it via integrating $$y = \sqrt{1 - x^2}; x={-1} to {1} $$
